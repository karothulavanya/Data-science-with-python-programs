```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
xpoints=np.array([1,2])
ypoints=np.array([1,100])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
plt.plot([5,200],[5,200],"--")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.scatter(x,y)
plt.show()
```

```python
xpoints=np.array([1,2,6,8])
ypoints=np.array([13,8,11,10])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
ypoints=np.array([13,4,8,9,2])
plt.plot(ypoints,marker="*",mec="r")
plt.show()
```

```python
ypoints=np.array([13,4,8,9,2])
plt.plot(ypoints,marker="o",mec="b",color="r")
plt.show()
```

```python
ypoints=np.array([13,4,8,9,2])
plt.plot(ypoints,marker="*",mec="r",linewidth="5")
plt.show()
```
