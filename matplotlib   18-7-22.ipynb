```python
#Bar graph
import matplotlib.pyplot as plt
country=["India","USA","Russia","France","Australia"]
GDP_per_capita=[48000,38000,35000,25000,15000]
plt.bar(country,GDP_per_capita)
plt.title('country vs GDP_per_capita')
plt.xlabel("country")
plt.ylabel("GDP per capita")
plt.show()
```

```python
import matplotlib.pyplot as plt
country=["India","USA","Russia","France","Australia"]
GDP_per_capita=[48000,38000,35000,25000,15000]
plt.barh(country,GDP_per_capita)
plt.title('country vs GDP_per_capita')
plt.xlabel("country")
plt.ylabel("GDP per capita")
plt.show()
```

```python
import matplotlib.pyplot as plt
country=["India","USA","Russia","France","Australia"]
GDP_per_capita=[48000,38000,35000,25000,15000]
plt.barh(country,GDP_per_capita,color="hotpink")
plt.title('country vs GDP_per_capita')
plt.xlabel("country")
plt.ylabel("GDP per capita")
plt.show()
```

```python
#Histogram
import matplotlib.pyplot as plt
numbers=[0.1,0.5,1,1.5,2,4,5.5,6,8,9]
plt.hist(numbers)
plt.xlabel("numbers")
plt.ylabel("Frequency")
plt.show()
```

```python
#pie charts
import matplotlib.pyplot as plot
import numpy as np
y=np.array([35,50,15])
co=["r","b","hotpink"]
la=["DSP","CD","WT"]
plt.pie(y,colors=co,labels=la)
plt.legend()
plt.show()
```
