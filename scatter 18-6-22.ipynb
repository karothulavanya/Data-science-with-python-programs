# Scatter

```python
import matplotlib.pyplot as plt
x=[4,6,8,2,5,7]
y=[10,20,30,40,50,60]
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[4,6,8,2,5]
y=[10,20,30,40,50]
colors=["red","pink","yellow","Black","Green"]
plt.scatter(x,y,c=colors)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[4,6,8,2,5]
y=[10,20,30,40,50]
plt.scatter(x,y,marker="^",c="red")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[4,6,8,2,5]
y=[10,20,30,40,50]
plt.scatter(x,y,marker=">",c="red")
plt.show()
```
